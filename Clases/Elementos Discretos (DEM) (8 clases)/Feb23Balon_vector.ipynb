```python
class Cuerpo:
    def __init__(self,x0,y0,z0,Vx0,Vy0,Vz0,m0):
        self.m = m0
        self.r = np.array([x0,y0,z0])
        self.V = np.array([Vx0,Vy0,Vz0])

    def CalculeFuerza(self):
        self.F = np.array([0,-self.m*g,0])
    def Muevase(self,dt):
        self.r = self.r + dt*self.V
        self.V = self.V + (dt/self.m)*self.F
```
