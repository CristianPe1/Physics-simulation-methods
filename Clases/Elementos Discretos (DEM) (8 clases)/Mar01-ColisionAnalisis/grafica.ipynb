```python
import pandas as pd
import matplotlib.pyplot as plt

def graficar(datos):

    path = "./"+ datos + ".txt"
    data = pd.read_csv(path,sep=' ',names=['N','Delta'])



    plt.plot(data['N'],data['Delta'],"-",label='Delta')
    plt.title('Error relativo en función de N')
    # plt.xscale("log")
    plt.yscale("log")
    plt.xlabel('N')
    plt.ylabel('log(Delta)')
    plt.savefig('mi_grafica.pdf')
    plt.grid()
    plt.legend()
```

```python
graficar("data")
```
