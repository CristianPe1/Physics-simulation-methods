# Cudratura de Gauss

<div class="mw-content-ltr mw-parser-output" lang="es" dir="ltr"><p>En <a href="/wiki/An%C3%A1lisis_num%C3%A9rico" title="Análisis numérico">análisis numérico</a> un método de cuadratura es una aproximación de una integral definida de una función. Una <b>cuadratura de Gauss</b> <i>n</i>, es una cuadratura construida para obtener el resultado exacto al integrar polinomios de grado 2n-1 o menos. Para esto selecciona los puntos de evaluación <i>x</i><sub>i</sub> y los pesos <i>w</i><sub>i</sub> de forma conveniente. La regla suele expresarse para una integral en el intervalo [−1, 1], y viene dada por la siguiente expresión:

```python
# Cuadratura de Gauss-Legendre

import numpy as np
import matplotlib.pyplot as plt

# Función a integrar
def f(x):
    return np.exp(-x**2)

# Cuadratura de Gauss-Legendre
def gauss_legendre(f, a, b, n):
    x, w = np.polynomial.legendre.leggauss(n)
    return (b-a)/2 * np.sum(w * f((b-a)/2 * x + (b+a)/2))

# Valores de n
n = np.arange(2, 10)

# Valores de la integral
I = np.zeros(len(n))
for i in range(len(n)):
    I[i] = gauss_legendre(f, -1, 1, n[i])

# Gráfica
plt.plot(n, I, 'o-')
plt.xlabel('n')
plt.ylabel('I')
plt.show()
```

```python
%%writefile Euler.cpp
#include <iostream>
#include <cmath>
using namespace std;

double f(double t, double x){
    return x/2;
}

void UnPasoEuler(double & t, double & x, double dt){
    double dx;
    dx = dt*f(t,x);
    x+=dx; t+=dt;
}

int main (void){
    double t,x;
    double dt=0.01;

    for(t=0,x=0.1;t<=10;){
        //cout<<t<<" "<<x<<endl;
        UnPasoEuler(t,x,dt);
    }

    cout<<x(<<""<<)x<<endl;
    cout<<x(<<""<<)<<x<<endl;

    return 0;

}
```
